```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import mne
from mne import create_info, EpochsArray, time_frequency
from mne.time_frequency import psd_array_welch
import seaborn as sns
```

```python
# CSVファイルからデータを読み込む
data = pd.read_csv('DP_droptime1.csv')
DMdata = data.drop('Marker', axis=1).values

# サンプリング周波数
sfreq = 256
n_channels = 14

# サンプリングレートに基づきエポックの長さを設定
epoch_length = sfreq  # 1秒

# データをエポックごとに整形
data_epochs = DMdata.reshape(-1, n_channels, epoch_length)

# MNE用の情報オブジェクトを作成
ch_names = ['ch'+str(i) for i in range(1, n_channels+1)]
ch_types = ['eeg'] * n_channels
info = create_info(ch_names=ch_names, ch_types=ch_types, sfreq=sfreq)

# データをエポックとしてロード
epochs = EpochsArray(data_epochs, info)

# 各チャンネルについてパワースペクトル密度（PSD）を計算してプロット
for i in range(n_channels):
    plt.figure(figsize=(10, 6))
    psds, freqs = time_frequency.psd_welch(epochs, picks=i)
    plt.semilogy(freqs, np.squeeze(psds))
    plt.title('Power Spectral Density (PSD) - Channel ' + str(i+1))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power Spectral Density (dB/Hz)')
    plt.show()
```

```python
# データとマーカーを分割
eeg_data = data.drop(columns=['Marker']).values
markers = data['Marker'].values

# サンプリング周波数
sfreq = 256
n_channels = 14

# サンプリングレートに基づきエポックの長さを設定
epoch_length = sfreq // 2  # 0.5秒

# データをエポックごとに整形
data_epochs = eeg_data.reshape(-1, n_channels, epoch_length)

# マーカーもエポックごとに整形
marker_epochs = markers.reshape(-1, epoch_length)

# マーカーに基づいて刺激前と刺激後のエポックを分割
pre_stimulus_epochs = data_epochs[marker_epochs[:, 0] == 0]
post_stimulus_epochs = data_epochs[marker_epochs[:, 0] == 1]

# MNE用の情報オブジェクトを作成
ch_names = ['ch'+str(i) for i in range(1, n_channels+1)]
ch_types = ['eeg'] * n_channels
info = create_info(ch_names=ch_names, ch_types=ch_types, sfreq=sfreq)

# 刺激前と刺激後のエポックをそれぞれロード
pre_epochs = EpochsArray(pre_stimulus_epochs, info)
post_epochs = EpochsArray(post_stimulus_epochs, info)

# 刺激前と刺激後のPSDを計算してプロット
for i in range(n_channels):
    plt.figure(figsize=(10, 6))
    
    # 刺激前のPSD
    psds, freqs = time_frequency.psd_array_welch(pre_epochs.get_data()[0, i], sfreq, fmin=1, fmax=40, n_fft=128, n_per_seg=128)
    plt.semilogy(freqs, np.squeeze(psds), label='Pre-stimulus')
    
    # 刺激後のPSD
    psds, freqs = time_frequency.psd_array_welch(post_epochs.get_data()[0, i], sfreq, fmin=1, fmax=40, n_fft=128, n_per_seg=128)
    plt.semilogy(freqs, np.squeeze(psds), label='Post-stimulus')

    plt.title('Power Spectral Density (PSD) - Channel ' + str(i+1))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power Spectral Density (dB/Hz)')
    plt.legend()
    plt.show()
```

```python
# CSVファイルからデータを読み込む
data = pd.read_csv('DP_all.csv')

# 刺激前と刺激後のデータに分ける
data_pre = data[data['Marker'] == 0]
data_post = data[data['Marker'] == 1]

# 各データをエポックとしてロード
sfreq = 256  # サンプリング周波数
n_channels = 14  # チャンネル数（Marker列を除く）

n_samples_per_epoch = int(sfreq * 0.5)  # 0.5秒のサンプル数
n_epochs_per_marker = data_pre.shape[0] // n_samples_per_epoch  # マーカーごとのエポック数

# 形状を (エポック数、チャンネル数、サンプル数) に変換
data_pre_epochs = data_pre.iloc[:, :-1].values.reshape(n_epochs_per_marker, n_channels, -1)
data_post_epochs = data_post.iloc[:, :-1].values.reshape(n_epochs_per_marker, n_channels, -1)

info = create_info(ch_names=[f'ch_{i+1}' for i in range(n_channels)], ch_types=['eeg'] * n_channels, sfreq=sfreq)

epochs_pre = EpochsArray(data_pre_epochs, info)
epochs_post = EpochsArray(data_post_epochs, info)

# 各周波数帯域での平均パワースペクトルを計算
freq_bands = {'Delta': [1, 4], 'Theta': [4, 8], 'Alpha': [8, 13], 'Beta': [13, 30]}
avg_psds_pre = {band: [] for band in freq_bands.keys()}
avg_psds_post = {band: [] for band in freq_bands.keys()}

n_fft = n_samples_per_epoch  # n_fftをエポックあたりのサンプル数と同じに設定

# データ処理のループ部分
for i in range(n_channels):
    band_psds_pre = []
    band_psds_post = []
    for band, freq_range in freq_bands.items():
        for epoch_pre, epoch_post in zip(epochs_pre, epochs_post):
            psds, freqs = psd_array_welch(epoch_pre, sfreq, fmin=freq_range[0], fmax=freq_range[1], n_fft=n_fft)
            print(psds.shape)
            print("Frequency bins:", freqs)
            band_psds_pre.append(psds.mean(axis=1))
            
            psds, freqs = psd_array_welch(epoch_post, sfreq, fmin=freq_range[0], fmax=freq_range[1], n_fft=n_fft)
            band_psds_post.append(psds.mean(axis=1))
        
    avg_psds_pre[band].append(np.mean(band_psds_pre))
    avg_psds_post[band].append(np.mean(band_psds_post))


# 各周波数帯域での平均PSDを計算
avg_psds_pre_means = {band: np.mean(psds) for band, psds in avg_psds_pre.items()}
avg_psds_post_means = {band: np.mean(psds) for band, psds in avg_psds_post.items()}

# 平均PSDをnumpy配列に変換
avg_psds_pre_array = np.array(list(avg_psds_pre_means.values()))
avg_psds_post_array = np.array(list(avg_psds_post_means.values()))

# 各周波数帯域での平均PSDの変化率を計算
change_rate = ((avg_psds_post_array - avg_psds_pre_array) / avg_psds_pre_array) * 100

# 幅と位置を設定
width = 0.35
ind = np.arange(len(freq_bands))

# プロットを作成
fig, ax = plt.subplots()
change_bars = ax.bar(ind, change_rate, width, label='Change Rate')

# ラベルを追加
ax.set_ylabel('Change Rate (%)')
ax.set_title('Change Rate of Average PSDs by Frequency Band')
ax.set_xticks(ind)
ax.set_xticklabels(freq_bands.keys())
ax.legend()

# プロットを表示
plt.show()
```

```python
# EPOC Xの14チャンネルを定義
ch_names = ['AF3', 'F7', 'F3', 'FC5', 'T7', 'P7', 'O1', 'O2', 'P8', 'T8', 'FC6', 'F4', 'F8', 'AF4']

# 刺激前と刺激後のデータを分割
data_before = data[data['Marker'] == 0][ch_names]
data_after = data[data['Marker'] == 1][ch_names]

# 各データフレームをMNEのRawオブジェクトに変換
raw_before = mne.io.RawArray(data_before.T, mne.create_info(ch_names, sfreq))
raw_after = mne.io.RawArray(data_after.T, mne.create_info(ch_names, sfreq))

# 特徴量を格納するためのデータフレームを作成
features_before = pd.DataFrame(index=ch_names)
features_after = pd.DataFrame(index=ch_names)

# 各チャンネルについてパワースペクトル密度（PSD）とバンドパワーを計算
for ch in ch_names:
    for raw, features in [(raw_before, features_before), (raw_after, features_after)]:
        psd, freqs = mne.time_frequency.psd_array_welch(raw.get_data(picks=ch), sfreq, fmin=0.5, fmax=45)
        psd = psd[0]

        # PSDを特徴量として追加
        features.loc[ch, 'PSD'] = np.mean(psd)

        # 各バンドのパワーを計算
        for band, freq_range in freq_bands.items():
            mask = np.logical_and(freqs >= freq_range[0], freqs <= freq_range[1])
            features.loc[ch, f'{band}Power'] = np.mean(psd[mask])

# 結果を表示
print("Before stimulus:")
print(features_before)

print("\nAfter stimulus:")
print(features_after)

# 特徴量のデータフレームを結合
features_before['Condition'] = 'Before'
features_after['Condition'] = 'After'
features = pd.concat([features_before, features_after]).reset_index().melt(id_vars=['index', 'Condition'], var_name='Feature', value_name='Value')

# バンドパワーのみを選択
df_bandpower = features[features['Feature'].str.contains('Power')]

# バンドパワーのプロット
plt.figure(figsize=(15, 10))
for i, ch in enumerate(ch_names, 1):
    plt.subplot(3, 5, i)
    df_bandpower_ch = df_bandpower[df_bandpower['index'] == ch]
    sns.barplot(x='Feature', y='Value', hue='Condition', ci=None, data=df_bandpower_ch)
    plt.title(ch)
    plt.ylim(0, df_bandpower['Value'].max() * 1.1)
    if i == 1:
        plt.legend()
    else:
        plt.legend().remove()
plt.tight_layout()
plt.show()
```
