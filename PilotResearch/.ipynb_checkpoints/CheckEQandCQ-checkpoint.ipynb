```python
import pandas as pd
import numpy as np
```

```python
# データを読み込む
df = pd.read_csv('MyEachDate/test_EPOCFLEX_8128_2023.06.26T17.12.49+09.00.csv', delimiter=',', skiprows=1)
df
```

```python
# EQ.OVERALLカラムがNaNではない最初の行を見つける
start_index = df['EQ.OVERALL'].first_valid_index()

# "CQ.Overall"の平均値を初めてEQ.OVERALLが出現するまで計算
avg_before_first_eq_overall = df['CQ.Overall'].iloc[0:start_index].mean()

# その行から64行ごとにスライス
sliced_df = df.loc[start_index::64]

# EQ.OVERALLが90以上の行を抽出
over_90 = sliced_df[sliced_df['EQ.OVERALL'] >= 90]

# 行の数をカウント
count = len(over_90)

print('EQ.OVERALLが90以上の行の数:', count)

# 90以上の値を持つ行とそれに対応するCQ.Overallの平均値を表示
for index, row in over_90.iterrows():
    if index+64 < len(df):
        avg = df['CQ.Overall'].iloc[index:index+64].mean()
    else:
        avg = df['CQ.Overall_END'].iloc[index:].mean()
    print(f'Index: {index-64}~{index}, EQ.OVERALL: {row["EQ.OVERALL"]}, CQ.Overall average: {avg}')
```
