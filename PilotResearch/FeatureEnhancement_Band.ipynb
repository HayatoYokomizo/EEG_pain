```python
import mne
import csv
import pandas as pd


#csv
R_csv = ["CloseEyes.csv", "CloseEyes_NoiseRemoval_A.csv", "OpenEyes.csv", "OpenEyes_NoiseRemoval_A.csv"]
S_csv = ["CloseEyes_Band.csv", "CloseEyes_NoiseRemoval_A_Band.csv", "OpenEyes_Band.csv", "OpenEyes_NoiseRemoval_A_Band.csv"]

# 
sfreq = 256

# 
bands = {'Delta': [0.5, 4],
         'Theta': [4, 8],
         'Alpha': [8, 13],
         'Beta': [13, 30]}

for i in range(len(R_csv)):
    # CSV
    data = pd.read_csv(R_csv[i])
    
    # 'timestamp'
    ch_names = [col for col in data.columns if col != 'timestamp']
    
    # 
    for ch in ch_names:
        # MNERaw
        raw = mne.io.RawArray([data[ch]], mne.create_info([ch], sfreq, ch_types='eeg'), verbose=False)
    
        # 
        for band, freq_range in bands.items():
            filtered_data = raw.copy().filter(l_freq=freq_range[0], h_freq=freq_range[1]).get_data()[0]
            data[f'{band}Filtered_{ch}'] = filtered_data
    
    
    # CSV
    data.to_csv(S_csv[i], index=False)
```
